```python
from river import tree
from river import metrics
from river import datasets
```

```python
# Create a Hoeffding Tree classifier
model = tree.HoeffdingTreeClassifier(
    grace_period=50,      # Number of instances before attempting a split
    delta=1e-5,           # Confidence level (Hoeffding bound)
    max_depth=10          # Maximum tree depth
)
```

```python
dataset = datasets.Phishing()
```

```python
metric = metrics.Accuracy()
```

```python
for i, (x, y) in enumerate(dataset):
    
    # Make a prediction BEFORE learning (prequential evaluation)
    y_pred = model.predict_one(x)
    
    # Update accuracy metric
    if y_pred is not None:
        metric.update(y, y_pred)
    
    # Train model with the current instance
    model.learn_one(x, y)
    
    # Print progress
    if i % 500 == 0:
        print(f"Processed {i} instances — Accuracy: {metric.get():.4f}")
```

```python
print("Final Accuracy:", metric.get())
```

```python
data_stream = [
    ({"attendance": 1, "quiz": 1}, "Pass"),
    ({"attendance": 0, "quiz": 0}, "Fail"),
    ({"attendance": 1, "quiz": 0}, "Pass"),
    ({"attendance": 1, "quiz": 1}, "Pass"),
    ({"attendance": 0, "quiz": 0}, "Fail"),
]

model = tree.HoeffdingTreeClassifier()
metric = metrics.Accuracy()

for x, y in data_stream:
    y_pred = model.predict_one(x)
    if y_pred:
        metric.update(y, y_pred)
    model.learn_one(x, y)

print("Accuracy:", metric.get())
```
