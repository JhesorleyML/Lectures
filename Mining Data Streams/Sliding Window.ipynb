```python
from collections import deque
```

```python
def stream_generator(n):
    for i in range(1, n + 1):
        yield i
```

```python
window_size = 10
```

```python
stream = stream_generator(50)
```

```python
window = deque(maxlen=window_size)
```

```python
for item in stream:
    window.append(item)
    print(f"Current Window: {list(window)}")
```
