```python
# Simulated streaming student activity data
student_stream = [
    {"login": 5, "minutes": 60, "quiz": 3},
    {"login": 1, "minutes": 10, "quiz": 0},
    {"login": 6, "minutes": 70, "quiz": 4},
    {"login": 2, "minutes": 15, "quiz": 1},
    {"login": 7, "minutes": 80, "quiz": 5},
    {"login": 3, "minutes": 30, "quiz": 2},
    {"login": 8, "minutes": 90, "quiz": 6},
    {"login": 1, "minutes": 8,  "quiz": 0},
]
```

```python
from river import cluster
```

```python
# STREAM-style clustering (one-pass, incremental)
stream_model = cluster.KMeans(
    n_clusters=3,   # Low, Medium, High activity
    halflife=0.4,   # Makes it adaptive to streams
    seed=42
)
```

```python
print("STREAM-style clustering results:\n")

for i, student in enumerate(student_stream):
    stream_model.learn_one(student)
    cluster_id = stream_model.predict_one(student)
    
    print(f"Student {i+1} {student} → Cluster {cluster_id}")
```
