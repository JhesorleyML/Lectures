```python
import pandas as pd
```

```python
data = {
    "Title": [
        "Book A", "Book B", "Book C", "Book D", "Book E",
        "Book F", "Book G", "Book H", "Book I", "Book J"
    ],
    "Year_Published": [
        2024, 2024, 2023, 2023, 2022,
        2021, 2020, 2019, 2018, 2017
    ]
}

books_df = pd.DataFrame(data)
```

```python
books_df
```

```python
recent_books = books_df[books_df["Year_Published"] >= 2023]
```

```python
recent_books
```

```python
books_df["Two_Year_Group"] = (books_df["Year_Published"] // 2) * 2

two_year_summary = books_df.groupby("Two_Year_Group").size().reset_index(name="Book_Count")
```

```python
two_year_summary
```

```python
books_df["Decade"] = (books_df["Year_Published"] // 10) * 10

decade_summary = (
    books_df
    .groupby("Decade")
    .size()
    .reset_index(name="Book_Count")
)
```

```python
decade_summary
```

```python
from IPython.display import display

print("RECENT YEARS (HIGH DETAIL)")
display(recent_books)

print("\nTWO-YEAR SUMMARY (MEDIUM DETAIL)")
display(two_year_summary)

print("\nDECADE SUMMARY (LOW DETAIL)")
display(decade_summary)
```

```python
hot_layer = books_df[books_df["Year_Published"] >= 2023]

print("CRITICAL / HOT LAYER (Recent Publications)")
display(hot_layer)
```

```python
books_df["Two_Year_Group"] = (books_df["Year_Published"] // 2) * 2

warm_layer = (
    books_df
    .groupby("Two_Year_Group")
    .size()
    .reset_index(name="Book_Count")
)

print("WARM LAYER (2-Year Aggregation)")
display(warm_layer)
```

```python
books_df["Decade"] = (books_df["Year_Published"] // 10) * 10

cold_layer = (
    books_df
    .groupby("Decade")
    .size()
    .reset_index(name="Book_Count")
)

print("COLD LAYER (Decade-Level Aggregation)")
display(cold_layer)
```

```python
layers_summary = pd.DataFrame({
    "Layer": ["Hot", "Warm", "Cold"],
    "Time Scope": ["Recent Years", "Medium-Term", "Long-Term"],
    "Granularity": ["Individual Books", "2-Year Groups", "Decades"],
    "Data Importance": ["High", "Medium", "Low"]
})

display(layers_summary)
```
