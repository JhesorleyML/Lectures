```python
import pandas as pd
```

```python
data = {
    "Brand": ["Honda", "Honda", "Honda", "Yamaha", "Yamaha", "Suzuki", "Suzuki"],
    "Year": [2022, 2023, 2024, 2023, 2024, 2022, 2023],
    "Model_Type": ["Scooter", "Scooter", "Underbone", "Scooter", "Big Bike", "Scooter", "Big Bike"],
    "Model_Count": [2, 3, 1, 2, 1, 1, 1]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#ROLL-UP (Total Models per Brand)
```

```python
roll_up = df.groupby("Brand")["Model_Count"].sum().reset_index()
roll_up
```

```python
#DRILL-DOWN (Brand → Brand + Year)
```

```python
drill_down = df.groupby(["Brand", "Year"])["Model_Count"].sum().reset_index()
drill_down
```

```python
#SLICE (Fix Year = 2023)
```

```python
slice_2023 = df[df["Year"] == 2023]
slice_2023
```

```python
#DICE (Brand = Honda or Yamaha AND Year = 2023–2024)
```

```python
dice_data = df[
    (df["Brand"].isin(["Honda", "Yamaha"])) &
    (df["Year"].isin([2023, 2024]))
]
dice_data
```

```python
#PIVOT (Brand vs Year)
```

```python
pivot_table = pd.pivot_table(
    df,
    values="Model_Count",
    index="Brand",
    columns="Year",
    aggfunc="sum",
    fill_value=0
)

pivot_table
```
