```python
from collections import Counter
```

```python
data_stream = [
    {"Engine Oil", "Spark Plug"},
    {"Engine Oil", "Air Filter"},
    {"Engine Oil", "Spark Plug"},
    {"Engine Oil", "Spark Plug"},
    {"Brake Pad", "Engine Oil"},
    {"Engine Oil", "Air Filter"},
    {"Engine Oil", "Spark Plug"}
]

data_stream
```

```python
window_size = 5   # last 5 transactions
```

```python
def frequent_items(stream, window_size, min_support):
    window = stream[-window_size:]
    item_counter = Counter()

    for transaction in window:
        for item in transaction:
            item_counter[item] += 1

    frequent = {
        item: count
        for item, count in item_counter.items()
        if count >= min_support
    }

    return frequent
```

```python
min_support = 5

frequent_patterns = frequent_items(data_stream, window_size, min_support)
frequent_patterns
```
